```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Collecting and cleaning raw per game data from 4 seaons
Raw_NBA_Data = pd.read_csv('NBA_Data.csv')
Raw_NBA_Data.head(24)
```

```python
Raw_NBA_Data = Raw_NBA_Data[Raw_NBA_Data['Rk'] != 'Rk']
Raw_NBA_Data.head(25)
```

```python
Raw_NBA_Data.tail(25)
```

```python
Raw_NBA_Data['Year'].unique()
```

```python
NBA_Data_SS = Raw_NBA_Data[['Player', 'Pos', 'Age', 'MP', 'PTS', 'Year']]
NBA_Data_SS.rename(columns = {"Pos": "Position", "MP": "Minutes Per Game", 'PTS':'Points Per Game'}, inplace = True)
NBA_Data_SS['Age'] = NBA_Data_SS['Age'].astype(int)
NBA_Data_SS['Minutes Per Game'] = NBA_Data_SS['Minutes Per Game'].astype(float)
NBA_Data_SS['Points Per Game'] = NBA_Data_SS['Points Per Game'].astype(float)
NBA_Data_SS = NBA_Data_SS[NBA_Data_SS['Minutes Per Game'] >= 10.0]
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['C'], 'Center')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['PF'], 'Power Foward')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['SF'], 'Small Foward')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['SG'], 'Shooting Guard')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['PG'], 'Point Guard')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['PG-SG'], 'Point Guard')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['SG-SF'], 'Shooting Guard')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['SF-PF'], 'Small Foward')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['PF-C'], 'Power Foward')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['PF-SF'], 'Power Foward')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['SF-SG'], 'Small Foward')
NBA_Data_SS['Position'] = NBA_Data_SS['Position'].replace(['C-PF'], 'Center')
NBA_Data_SS.head()
```

```python
#Creating the interactive visualization
min_x,max_x = NBA_Data_SS['Age'].min(), NBA_Data_SS['Age'].max()
min_y,max_y = NBA_Data_SS['Points Per Game'].min(), NBA_Data_SS['Points Per Game'].max()
min_x = min_x - 0.1*(max_x - min_x)
min_y = min_y - 0.1*(max_y - min_y)
max_x = max_x + 0.1*(max_x - min_x)
max_y = max_y + 0.1*(max_y - min_y)

fig = px.scatter(NBA_Data_SS, x = 'Age', y = 'Points Per Game', hover_name = 'Player', color = 'Position',
                 hover_data = {'Position': False, 'Year': False, 'Minutes Per Game': True}, animation_frame = 'Year', 
                 color_discrete_sequence = ["red", "yellow", "green", "blue", "magenta"],
                 height = 700, width = 1300, range_x = [min_x, 43], range_y= [0, 40],
                 title = 'The Affect of Age on an NBA Players Point Per Game (1985-2016)')

fig.update_layout(sliders=[{"currentvalue": {"prefix": "Year: ",'xanchor':'right'}}])
fig.update_layout(title_font_family = "Times New Roman")
fig.update_layout(title = {'y': 0.9, 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top'})
fig.update_layout(title = dict(font = dict(size = 30)))
fig.update_traces(marker = {'size':15, 'opacity':0.7})
fig.update_traces(marker = dict(line = dict(width = 1.5, color = 'black')))
fig.write_html('Players_Age_VS_Points_inter.html')
fig.write_image('Players_Age_VS_Points_inter.png')
fig.show()
```
